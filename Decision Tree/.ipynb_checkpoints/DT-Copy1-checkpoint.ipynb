# Decision Tree Classification

## Importing the libraries

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree   
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```python
Person = pd.read_csv('DatasetFinal.csv')
print(type(Person))
print(Person)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
Person.columns
```

```python
x=Person.loc[:,['Sex', 'Marital status', 'Age', 'Experience']]
#x['LFDeficient']=pd.Categorical(Person['LFDeficient']).codes
y=Person.loc[:,['Membership']]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

## Training the Decision Tree Classification model on the Training set

```python
classifier = DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=1)
classifier.fit(x_train,y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(x_test)
print('The accuracy of the Decision Tree is','{:.3f}'.format(100*metrics.accuracy_score(y_pred,y_test)))
```
