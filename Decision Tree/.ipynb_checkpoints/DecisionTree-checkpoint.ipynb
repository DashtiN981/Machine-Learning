```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree   
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
weather=pd.read_excel('weather conditions.xlsx')
weather
```

```python
weather.columns
```

```python
x=weather[['Outlook', 'Temperature', 'Humidity', 'Windy']]
y=pd.Categorical(weather['Play Golf']).codes
y
```

```python
import numpy as np
def entropy2(p):
    return - p*np.log2(p) - (1 - p)*np.log2((1 - p))

print(y[(y > 0)].shape)
print(entropy2(9/14))
print(entropy2(5/14))
print(entropy2(7/14))
print(entropy2(14/14))
```

```python
weather[[ 'Temperature', 'Humidity', 'Windy','Play Golf']].sort_values(by='Play Golf')
```

```python
x.loc[:,('Outlook')] = x['Outlook'].astype('category')
x.loc[:,('Outlook')] = x['Outlook'].cat.codes

x.loc[:,('Temperature')] = x['Temperature'].astype('category')
x.loc[:,('Temperature')] =x['Temperature'].cat.codes

x.loc[:,['Humidity']] = x['Humidity'].astype('category')
x.loc[:,['Humidity']] =x['Humidity'].cat.codes

x.loc[:,['Windy']] = x['Windy'].astype('category')
x.loc[:,['Windy']] =x['Windy'].cat.codes
x
```

```python
y
```

```python
mod_dt = DecisionTreeClassifier(
                                    criterion='entropy',
                                    max_depth = 2,
                                    random_state = 100
                                )
mod_dt.fit(x,y)
prediction=mod_dt.predict(x)

print(metrics.accuracy_score(prediction,y))
```

```python
plt.figure(figsize = (33,10))
plot_tree(mod_dt, feature_names = x.columns, class_names = ['no','yes'], filled = True,max_depth=8);
```

```python
pd.concat([x,weather['Play Golf']],axis=1).sort_values(by=['Outlook']).sort_values(by=['Humidity'])
```

```python
iris = pd.read_csv('Dataset/iris.csv')
print(type(iris))
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris.columns
```

```python
x=iris.loc[:,['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y=iris.loc[:,['class']]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
dt=DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=1)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print('The accuracy of the Decision Tree is','{:.3f}'.format(100*metrics.accuracy_score(y_pred,y_test)))
```

```python
iris.groupby('class').all()
```

```python
plt.figure(figsize = (33,20))
plot_tree(dt,    proportion=False, feature_names = x.columns,class_names = ['Iris-setosa','Iris-versicolor','Iris-virginica'], filled = True,max_depth=10);
```

```python
titanic=pd.read_excel('Dataset/titanic3.xlsx')
```

```python
titanic.head()
```

```python
tit2=titanic[['pclass','sex','age','sibsp','survived']]
```

```python
tit2.isna().sum()
```

```python
tit2=tit2.dropna()
tit2.isna().sum()
```

```python
tit2.describe()
```

```python
tit2.columns
```

```python
tit2['sex']=tit2['sex'].replace({'male':100,'female':200})
```

```python
y=tit2['survived']
x=tit2[['pclass', 'sex', 'age', 'sibsp']]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
dt=DecisionTreeClassifier(criterion='entropy',max_depth=10,random_state=1)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print('The accuracy of the Decision Tree is','{:.3f}'.format(100*metrics.accuracy_score(y_pred,y_test)))
```

```python
plt.figure(figsize = (33,20))
plot_tree(dt,    proportion=False, feature_names = x.columns, filled = True,max_depth=3);
```
